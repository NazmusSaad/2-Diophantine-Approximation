```python
# NEW GOOD COPY
# LLL MATRIX Generating code --> Hardcode for each use case for now becuase the primes can change which means different initial matrix
# Step 0: Precomputation

import numpy as np
import sympy as sp

# Define the two prime numbers
primes = [2, 3]

# Compute the logarithms of the primes
log_primes = [sp.log(p).evalf() for p in primes]

# Construct the 3x2 matrix
initial_matrix = np.array([
    [-2 * log_primes[0], log_primes[1]],
    [-3 * log_primes[0], 2 * log_primes[1]],
    [19 * log_primes[0], -12 * log_primes[1]]
], dtype=np.float64)

# # Construct the 3x2 matrix
# initial_matrix = np.array([
#     [-2, 1],
#     [-3, 2 ],
#     [19, -12 ]
# ], dtype=np.float64)

print("Initial Matrix:")
print(initial_matrix)

def gram_schmidt(B):
    """Performs Gram-Schmidt orthogonalization."""
    n = B.shape[0]  # Number of rows
    m = B.shape[1]  # Number of columns
    U = np.zeros_like(B)
    R = np.zeros((m, m), dtype=np.float64)  # Square matrix R of size m x m

    for i in range(n):
        U[i] = B[i]
        for j in range(min(i, m)):  # Ensure j does not exceed the number of columns
            R[j, min(i, m-1)] = np.dot(U[j], B[i]) / np.dot(U[j], U[j])
            U[i] -= R[j, min(i, m-1)] * U[j]
        if i < m:
            R[i, i] = np.linalg.norm(U[i])

        # If U[i] contains NaN, exit early to prevent infinite loop
        if np.isnan(U[i]).any():
            print(f"Warning: NaN detected in Gram-Schmidt orthogonalization at row {i}.")
            break

    return U, R

def lll_reduction(B, delta=0.75):
    """Applies the LLL algorithm to matrix B."""
    n, m = B.shape
    U, R = gram_schmidt(B)

    k = 1
    max_iterations = 100  # Set a maximum number of iterations to avoid infinite loop
    iteration_count = 0

    while k < n and iteration_count < max_iterations:
        print(f"Iteration {k}:")
        print("Current B matrix:")
        print(B)
        print("Current U matrix:")
        print(U)

        # Size reduction
        for j in range(min(k - 1, m - 1), -1, -1):  # Adjusted range to avoid index errors
            mu = np.dot(B[k], U[j]) / np.dot(U[j], U[j])
            if abs(mu) > 0.5:
                B[k] -= np.round(mu) * B[j]

        U, R = gram_schmidt(B)

        # Check for NaN values in U matrix
        if np.isnan(U).any():
            print("Stopping due to NaN values in U matrix.")
            break

        # Lovasz condition, adjusted for rectangular matrices
        if np.dot(U[k], U[k]) >= (delta - (np.dot(B[k], U[k - 1]) / np.dot(U[k - 1], U[k - 1]))**2) * np.dot(U[k - 1], U[k - 1]):
            k += 1
        else:
            B[[k, k - 1]] = B[[k - 1, k]]
            U, R = gram_schmidt(B)
            k = max(k - 1, 1)

        iteration_count += 1

    if iteration_count >= max_iterations:
        print("Max iterations reached. Exiting to prevent infinite loop.")

    return B

# Apply LLL reduction
lll_reduced_matrix = lll_reduction(initial_matrix)

# Display the LLL-reduced matrix
print("LLL-reduced d_ij matrix:")
print(lll_reduced_matrix)
```

```python
# factorial
def fact(n):
  if n == 0 or n == 1:
    return 1
  return n*fact(n-1)
```

```python
import math
# Step 3: Evaluate e^ei using EXP

# EXP : Taylor series expansion of exp(x) series then evaluate using rectangular splitting

def EXP(e):
  N = 100 # number of terms in the series
  p = N//5
  q = N // p
  sum, sum2 = 0, 0

  # Normal
  for k in range(N):
    sum += (math.pow(e, k)/fact(k))

  # Rectangular splitting
  for i in range(p - 1):
    for j in range(q - 1):
        sum2 += (math.pow(e, (p * j + i)) / math.factorial(p * j + i))

  # print(sum, sum2)
  return sum
```
